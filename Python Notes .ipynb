```python
#10

# Use python to produce code below that will perform the following:

# Read multiple lines from the user on standard input until an empty string is read.
# Return a list of all these lines without line terminators
# Each line should be reversed from how it is read in.

def reverseit():
    user_list = []
    user_input = input()
    while user_input:
        user_list.append(user_input[::-1])
        user_input = input()
    return user_list

```

```python
#10

# Given a mixed case string as parameter s
# Capitalize every letter with an even index within the string.
# Lowercase every letter with an odd index within the string.
# Return the resulting string.
#Example - Given: "ABCDEF ghijkl" Return: "AbCdEf gHiJkL"

s = 'ABCDEF ghijkl'

def leetString(s):
    i=0
    new_string = ''
    for c in s:
        if i%2:
            new_sting = new_string+c.lower()
        else:
            new_string = new_string+c.upper()
    i = i+1
    return(new_sting)
```

```python
#10 (Alternate)

s = 'ABCDEF ghijkl'
def leetString(s):
    new_string = ''
    for i in range(len(s)):
        if i%2:
            new_string = new_string + s[i].lower()
    else:
        new_string = new_string + s[i].upper()
    return(new_string)
```

```python
#ooB

# Modify code below and implement guess_number so that it repeatedly asks the user for a number between 0 and 100, inclusive. 
# If the user correctly guesses the value of the given argument n, print 'WIN' and return. 
# Otherwise, indicate whether the guess was too high or too low.


def guess_number(n):
    #n = range(0,101)
    user_input = int(input('Give me a number between 0-100'))
    while user_input:
        if user_input == n:
            print('WIN')
            break
        elif user_input > n:
            print('too high')
            user_input = int(input('Give me another number between 0-100'))
        else:
            print('too low')
            user_input = int(input('Give me another number between 0-100'))

guess_number(23)
```
